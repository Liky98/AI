```python
video_file = "./sample/Mov01.mp4"

cap = cv2.VideoCapture(video_file) #

if cap.isOpened() :
    while True :
        ret, img0 = cap.read() # 
        if ret :#  
            cv2.imshow(video_file, img0)
            if cv2.waitKey(25) != -1 : #waitKey  -1 
                break
                #cv2.waitKey(325) #25ms
        else : break
else : 
    print("can't open video")

cap.release() # 
cv2.destroyAllWindows()

#(ms) = 1000(1s) / fps
```

# window setting

```python

file_path = "./sample/test2.jpg"
img = cv2.imread(file_path)                            #    
img_gray = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE) #    

cv2.namedWindow('origin')                               # origin   
cv2.namedWindow('gray', cv2.WINDOW_NORMAL)              # gray   
cv2.imshow('origin', img)                               # origin   
cv2.imshow('gray', img_gray)                            # gray   

cv2.moveWindow('origin', 0, 0)                          #   
cv2.moveWindow('gray', 100, 100)                        #   

cv2.waitKey(0)                                          #  
cv2.resizeWindow('origin', 200, 200)                    #    ( )
cv2.resizeWindow('gray', 100, 100)                      #    ( ))

cv2.waitKey(0)                                          #  
cv2.destroyWindow("gray")                               # gray  
cv2.waitKey(0)                                          #  
cv2.destroyAllWindows()                                 #   
```

# Key event

```python
  
import cv2
import numpy as np

img_file = file_path
img = cv2.imread(img_file) 
title = 'IMG'                   #   
x, y = 100, 100                 #  

while True:
    cv2.imshow(title, img)
    cv2.moveWindow(title, x, y)
    key = cv2.waitKey(0) & 0xFF #    , 8 
    print(key, chr(key))        #   ,    
    if key == ord('h'):         # 'h'    
        x -= 10
    elif key == ord('j'):       # 'j'    
        y += 10
    elif key == ord('k'):       # 'k'    
        y -= 10
    elif key == ord('l'):       # 'l'    
        x += 10
    elif key == ord('q') or key == 27: # 'q'  'esc'  
        break
        cv2.destroyAllWindows()
    cv2.moveWindow(title, x, y )   #    
        
```

```python
import cv2
import numpy as np

win_name = 'Trackbar'                                   #  

img = cv2.imread('./sample/DrawSample1.jpg')
cv2.imshow(win_name,img)                                #    

#      ---
def onChange(x):                                        
    print(x)                                            #     --- 
    # 'R', 'G', 'B'        --- 
    r = cv2.getTrackbarPos('R',win_name)               
    g = cv2.getTrackbarPos('G',win_name)               
    b = cv2.getTrackbarPos('B',win_name)               
    print(r, g, b)
    img[:] = [b,g,r]                                    #       --- 
    cv2.imshow(win_name, img)                           #    

#      --- 
cv2.createTrackbar('R', win_name, 255, 255, onChange)  
cv2.createTrackbar('G', win_name, 255, 255, onChange)
cv2.createTrackbar('B', win_name, 255, 255, onChange)

while True:
    if cv2.waitKey(1) & 0xFF == 27:
        break
cv2.destroyAllWindows()        
```
